```python
# import dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

# Store CSV into DataFrame

```python
# import the ratings CSV file
rating_file = "Resources/all_games_ratings.csv"
ratings_df = pd.read_csv(rating_file)
ratings_df
```

# Create new data with select columns

```python
# Create a filtered dataframe from specific columns
ratings_cols = ["name", "platform", "release_date", "user_review"]
ratings_transformed = ratings_df[ratings_cols].copy()
ratings_transformed.head()
```

# Transform the data from the 'Name' column

```python
# Transform the name of the games to be able to merge later
ratings_transformed['name'] = ratings_transformed['name'].str.replace(" ", "_")
ratings_transformed['name'] = ratings_transformed['name'].str.replace(":", "")
ratings_transformed['name'] = ratings_transformed['name'].str.replace("'", "")
ratings_transformed['name'] = ratings_transformed['name'].str.lower()
ratings_transformed
```

# Store CSV into DataFrame

```python
# import the video game sales CSV file
sales_file = "Resources/video_game_sales.csv"
sales_df = pd.read_csv(sales_file)
sales_df
```

# Create new data with select columns

```python
# Create a filtered dataframe from specific columns
sales_cols = ["Name", "Global_Sales"]
sales_transformed = sales_df[sales_cols].copy()
sales_transformed.head()
```

# Transform the data from the 'Name' column

```python
# Transform the name of the games to be able to merge later
sales_transformed['Name'] = sales_transformed['Name'].str.replace(" ", "_")
sales_transformed['Name'] = sales_transformed['Name'].str.replace(":", "")
sales_transformed['Name'] = sales_transformed['Name'].str.replace("'", "")
sales_transformed['Name'] = sales_transformed['Name'].str.replace(".", "")
sales_transformed['Name'] = sales_transformed['Name'].str.replace("/", "_")
sales_transformed['Name'] = sales_transformed['Name'].str.replace("!", "")
sales_transformed['Name'] = sales_transformed['Name'].str.lower()
sales_transformed.head()
```

# Rename the column titles

```python
sales_transformed = sales_transformed.rename(columns={"Name": "name",
                                                      "Global_Sales": "global_sales"})
sales_transformed
```

# Merge the two DataFrames together

```python
# Merging the two dataframes together
merge_df = pd.merge(ratings_transformed, sales_transformed, on="name", how="inner")
merge_df
```

# Drop Duplicates from the new DataFrame

```python
# Dropping duplicates from the merged dataframe
merge_df.drop_duplicates("name", inplace=True)
```

```python
# checking the datatypes for each column in the new dataframe
merge_df.dtypes
```

```python
# changing the datatype for the column "user_review"
merge_df["user_review"] = merge_df["user_review"].astype(float)
merge_df.dtypes
```

# Create connection to the Database

```python
# create the connection to the relational database
rds_connection_string = "postgres:postgres@localhost:5432/etl_project2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# inspect the tables created within the database
inspector = inspect(engine)
print(inspector.get_table_names())
```

# Use pandas to load merged DataFrame into relational database

```python
# load the data into the database using pandas
merge_df.to_sql(name='video_games', con=engine, if_exists='append', index=False)
```

# Confirm data has been added by querying the table

```python
# confirming that the data has been sent to database
pd.read_sql_query('select * from video_games', con=engine).head()
```
